```python
# Task 1

filename = "python_notes.txt"

# Part A: Write and Append
notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes."
]

# Writing initial lines
with open(filename, "w", encoding="utf-8") as f:
    for line in notes:
        f.write(line + "\n")
print(f"✓ Initial notes written to {filename} successfully.")

# Appending two custom lines
custom_notes = [
    "Topic 6: File I/O allows persistent data storage.",
    "Topic 7: APIs enable communication between different software systems."
]
with open(filename, "a", encoding="utf-8") as f:
    for line in custom_notes:
        f.write(line + "\n")
print("✓ Custom lines appended successfully.")

# Part B: Read and Search
print("\n--- Reading File Content ---")
line_count = 0
with open(filename, "r", encoding="utf-8") as f:
    for i, line in enumerate(f, 1):
        print(f"{i}. {line.strip()}")
        line_count = i

print(f"\nTotal number of lines in file: {line_count}")

# Keyword Search
keyword = input("\nEnter a keyword to search for in the notes: ").lower()
found = False
with open(filename, "r", encoding="utf-8") as f:
    for line in f:
        if keyword in line.lower():
            print(f"Match found: {line.strip()}")
            found = True

if not found:
    print(f"No lines found containing the keyword: '{keyword}'")
```

```python
# Task 2 API integration

import requests

# Base URL for the DummyJSON Products API
BASE_URL = "https://dummyjson.com/products"

# --- Step 1: Fetch and Display 20 Products ---
print("--- Step 1: Fetching 20 Products ---")
response = requests.get(f"{BASE_URL}?limit=20")
data = response.json()
products = data['products']

# Print the formatted table
print(f"{'ID':<4} | {'Title':<30} | {'Category':<15} | {'Price':<8} | {'Rating'}")
print("-" * 75)
for p in products:
    # We use string slicing [:30] to ensure long titles don't break the table alignment
    print(f"{p['id']:<4} | {p['title'][:30]:<30} | {p['category']:<15} | ${p['price']:<7} | {p['rating']}")


# --- Step 2: Filter and Sort ---
print("\n--- Step 2: Filtering (Rating >= 4.5) and Sorting (Price Descending) ---")
filtered_products = [p for p in products if p['rating'] >= 4.5]

# Sorting the filtered list by price in descending order
sorted_products = sorted(filtered_products, key=lambda x: x['price'], reverse=True)

for p in sorted_products:
    print(f"Product: {p['title']:<25} | Price: ${p['price']:<7} | Rating: {p['rating']}")


# --- Step 3: Search by Category (Laptops) ---
print("\n--- Step 3: Laptops Category ---")
laptop_res = requests.get(f"{BASE_URL}/category/laptops")
laptops = laptop_res.json()['products']

for laptop in laptops:
    print(f"Laptop: {laptop['title']} - Price: ${laptop['price']}")


# --- Step 4: POST Request (Simulated) ---
print("\n--- Step 4: Simulating POST Request ---")
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

post_response = requests.post(f"{BASE_URL}/add", json=new_product)
print("Server Response:")
print(post_response.json())
```

```python
# Task 3 exception handling

import requests

# --- Part A: Guarded Calculator ---
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("--- Part A: Testing safe_divide ---")
print(f"10 / 2    -> {safe_divide(10, 2)}")
print(f"10 / 0    -> {safe_divide(10, 0)}")
print(f"'ten' / 2 -> {safe_divide('ten', 2)}")


# --- Part B: Guarded File Reader ---
def read_file_safe(filename):
    try:
        with open(filename, "r") as f:
            content = f.read()
            return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None
    finally:
        # This block ALWAYS runs, even if the file is missing
        print("File operation attempt complete.")

print("\n--- Part B: Testing read_file_safe ---")
# This should work if you already ran Task 1
read_file_safe("python_notes.txt")
# This should trigger the Error message
read_file_safe("ghost_file.txt")


# --- Part C: Robust API Calls ---
# Here we add timeouts and network error handling to the API logic
print("\n--- Part C: Robust API Call Test ---")
try:
    # Adding a timeout of 5 seconds as per requirements
    response = requests.get("https://dummyjson.com/products?limit=5", timeout=5)
    response.raise_for_status()
    print("API Call Successful!")
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")


# --- Part D: Input Validation Loop ---
print("\n--- Part D: Product Lookup Loop ---")
while True:
    choice = input("Enter a product ID to look up (1–100), or 'quit' to exit: ").strip().lower()

    if choice == 'quit':
        print("Exiting lookup system.")
        break

    # Validation before API call
    if not choice.isdigit():
        print("Warning: Please enter a numeric ID.")
        continue

    prod_id = int(choice)
    if not (1 <= prod_id <= 100):
        print("Warning: ID must be between 1 and 100.")
        continue

    # Make the API call only after validation
    try:
        res = requests.get(f"https://dummyjson.com/products/{prod_id}", timeout=5)

        if res.status_code == 200:
            product_data = res.json()
            print(f"Success! Item: {product_data['title']} | Price: ${product_data['price']}")
        elif res.status_code == 404:
            print("Product not found.")

    except Exception as e:
        print(f"Network error occurred: {e}")
```
